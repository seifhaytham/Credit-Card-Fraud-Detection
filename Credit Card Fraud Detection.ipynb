```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("creditcard_2023.csv")
df.head()
```

```python
df = df.drop('id', axis=1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates()
```

```python
target = 'Class'
corr = df.corr()
target_corr = corr[target].abs()
threshold = 0.1
drop = target_corr[target_corr < threshold].index.tolist()
if target in drop:
    drop.remove(target)
df = df.drop(columns=drop)
```

```python
df.corr()
```

```python
sns.countplot(x='Class', data=df)
plt.title('Class Distribution')
plt.show()
```

```python
X = df.drop(target , axis = 1)
Y = df[target]
```

```python
scalar = StandardScaler()
X = scalar.fit_transform(X)
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train , y_train)
```

```python
y_pred1 = LR.predict(X_test)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
report = classification_report(y_test,y_pred1)
print(f'The classification report \n{report}')
```

```python
conf = confusion_matrix(y_test,y_pred1)
sns.heatmap(conf,annot=True,cmap="Blues",fmt='d')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# XGBoost

```python
from xgboost import XGBClassifier
XG = XGBClassifier()
XG.fit(X_train , y_train)
```

```python
y_pred2 = XG.predict(X_test)
y_pred2
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
report = classification_report(y_test,y_pred2)
print(f'The classification report \n{report}')
```

```python
conf = confusion_matrix(y_test,y_pred2)
sns.heatmap(conf,annot=True,cmap="Blues",fmt='d')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
